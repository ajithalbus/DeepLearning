```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_learning_curve(title, X):
    plt.figure()
    plt.title(title)
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    epochs=range(20)
    sizes=['Squared-Error','Cross-Entropy']
    colors=['r','b','k','g']
    plt.grid()
    j=0
    plt.xticks(epochs)
    for i in X:
        plt.plot(epochs,i, '.-',color=colors[j],
             label=str(sizes[j]))
        j+=1
    plt.legend(loc="best")
    return plt


```
